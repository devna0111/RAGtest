```python
%%time    
import pandas as pd
from openpyxl import load_workbook
from PIL import Image
import io, base64, requests, os, logging

logging.basicConfig(
    level=logging.INFO,
    format='[%(asctime)s] %(message)s',
    datefmt='%H:%M:%S'
)

def extract_text_and_table(file_path):
    xls = pd.ExcelFile(file_path)
    results = []

    for sheet in xls.sheet_names:
        df = pd.read_excel(xls, sheet_name=sheet, dtype=str)
        flat_text = sorted(set(val.strip() for val in df.astype(str).stack() if val and val.strip()))
        table_markdown = df.to_markdown(index=False)

        summary_lines = []
        table_type = ""
        description_lines = []

        numeric_df = df.apply(pd.to_numeric, errors='coerce')
        numeric_cols = [col for col in numeric_df.columns if numeric_df[col].notna().sum() > 0]
        shape_desc = f"-  í…Œì´ë¸” í˜•íƒœ: {df.shape[0]}í–‰ Ã— {df.shape[1]}ì—´"
        summary_lines.append(shape_desc)

        # í…Œì´ë¸” ìœ í˜•
        if "Department" in df.columns:
            table_type = "ë¶€ì„œë³„ ë¹„êµ í…Œì´ë¸” (ë²”ì£¼í˜• í–‰ vs ìˆ˜ì¹˜í˜• ì—´)"
        elif df.dtypes[0] == 'object':
            table_type = "ë²”ì£¼í˜• í‚¤ì— ìˆ˜ì¹˜ ê°’ì„ ê°€ì§„ í…Œì´ë¸”"
        else:
            table_type = "ìˆ˜ì¹˜ ì¤‘ì‹¬ì˜ í–‰ë ¬ ë˜ëŠ” ì¸¡ì • í…Œì´ë¸”"

        summary_lines.append(f"-  í…Œì´ë¸” ìœ í˜•: {table_type}")

        if numeric_cols:
            for col in numeric_cols:
                max_val = numeric_df[col].max()
                avg_val = numeric_df[col].mean()
                summary_lines.append(f"-  ì—´ '{col}': ìµœëŒ€ê°’={max_val}, í‰ê· ={avg_val:.2f}")

        # í–‰ë³„ ìš”ì•½
        if len(numeric_cols) >= 1:
            key_col = df.columns[0]
            for idx, row in df.iterrows():
                parts = [f"{row[key_col]}:"]
                for col in numeric_cols:
                    val = row[col]
                    if pd.notna(val):
                        parts.append(f"{col} {val}")
                description_lines.append(" / ".join(parts))

        results.append({
            "sheet": sheet,
            "text": flat_text,
            "text_summary": f"- ì´ ì‹œíŠ¸ì—ëŠ” ì´ {len(flat_text)}ê°œì˜ ê³ ìœ  í…ìŠ¤íŠ¸ í•­ëª©ì´ ìˆìŠµë‹ˆë‹¤.",
            "table_type": table_type,
            "table_summary": "\n".join(summary_lines),
            "table_description": "\n".join(description_lines),
            "table_markdown": table_markdown
        })

    return results

def extract_images_from_excel(file_path):
    wb = load_workbook(file_path)
    imgs = []
    for sheet in wb.worksheets:
        for img in getattr(sheet, "_images", []):
            if hasattr(img, "_data"):
                img_bytes = img._data()
                pil = Image.open(io.BytesIO(img_bytes)).convert("RGB")
                pil.thumbnail((640, 480))
                imgs.append({"sheet": sheet.title, "image": pil})
    return imgs

def run_qwen_vl_on_chart_image(pil_image):
    buf = io.BytesIO()
    pil_image.save(buf, format="PNG")
    img_b64 = base64.b64encode(buf.getvalue()).decode("utf-8")

    prompt = (
        "ì´ ì´ë¯¸ì§€ëŠ” ì—‘ì…€ ë¬¸ì„œì— í¬í•¨ëœ ì°¨íŠ¸ ë˜ëŠ” í‘œ ì´ë¯¸ì§€ì…ë‹ˆë‹¤.\n"
        "ë‹¤ìŒ í•­ëª©ì„ í•œêµ­ì–´ë¡œ ì •í™•í•˜ê²Œ ì„¤ëª…í•´ì£¼ì„¸ìš”:\n"
        "1. ì°¨íŠ¸ë‚˜ í‘œì˜ ì œëª© ë˜ëŠ” ì£¼ì œëŠ” ë¬´ì—‡ì¸ê°€ìš”?\n"
        "2. Xì¶•ê³¼ Yì¶•ì€ ê°ê° ë¬´ì—‡ì„ ì˜ë¯¸í•˜ë‚˜ìš”?\n"
        "3. Xì¶• í•­ëª©ë³„ë¡œ ìƒì‘í•˜ëŠ” Yê°’(ìˆ«ì)ì„ êµ¬ì²´ì ìœ¼ë¡œ ë‚˜ì—´í•´ì£¼ì„¸ìš”.\n"
        "4. ëˆˆì— ë„ëŠ” ìˆ˜ì¹˜ë‚˜ íŒ¨í„´ì€ ë¬´ì—‡ì¸ê°€ìš”?\n"
        "5. ì „ì²´ ë‚´ìš©ì„ í•œë‘ ë¬¸ì¥ìœ¼ë¡œ ìš”ì•½í•´ì£¼ì„¸ìš”."
    )

    res = requests.post(
        "http://localhost:11434/api/generate",
        json={
            "model": "qwen2.5vl",
            "prompt": prompt,
            "images": [img_b64],
            "stream": False
        }
    )
    if res.status_code == 200:
        return res.json().get("response", "").strip()
    else:
        raise RuntimeError(f"Ollama í˜¸ì¶œ ì‹¤íŒ¨: {res.status_code}")

def assemble_excel_report(file_path):
    text_table = extract_text_and_table(file_path)
    images = extract_images_from_excel(file_path)

    report = ""

    for sheet_info in text_table:
        name = sheet_info["sheet"]
        report += f"\n\n## ğŸ“„ ì‹œíŠ¸: {name}\n"
        report += "\n### ğŸ“ í…ìŠ¤íŠ¸ í•­ëª©\n"
        report += "\n".join(f"- {t}" for t in sheet_info["text"])
        report += "\n\n###  í…ìŠ¤íŠ¸ ìš”ì•½\n" + sheet_info["text_summary"]
        report += "\n\n###  í…Œì´ë¸” ìœ í˜•\n" + sheet_info["table_type"]
        report += "\n\n###  í…Œì´ë¸” ìš”ì•½\n" + sheet_info["table_summary"]
        report += "\n\n###  í–‰ë³„ ì„¤ëª…\n" + sheet_info["table_description"]
        report += "\n\n###  ë§ˆí¬ë‹¤ìš´ í…Œì´ë¸”\n" + sheet_info["table_markdown"]

    for i, img in enumerate(images):
        logging.info(f" ì´ë¯¸ì§€ {i+1}/{len(images)} í•´ì„ ì¤‘...")
        try:
            analysis = run_qwen_vl_on_chart_image(img["image"])
            report += f"\n\n##  ì°¨íŠ¸/í‘œ ì´ë¯¸ì§€ {i+1} (ì‹œíŠ¸: {img['sheet']})\n{analysis}"
        except Exception as e:
            report += f"\n\n##  ì´ë¯¸ì§€ í•´ì„ ì‹¤íŒ¨ {i+1}: {e}"

    return report.strip()

if __name__ == "__main__":
    file_path = r"C:\Ai_x\source\í”„ë¡œì íŠ¸2\sample_inputs\sample.xlsx"
    if os.path.exists(file_path):
        logging.info(f" ë¶„ì„ ì‹œì‘: {file_path}")
        print(assemble_excel_report(file_path))
    else:
        logging.error(" íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
```

```python
%%time
import pandas as pd
from openpyxl import load_workbook
from PIL import Image
import io, base64, logging

#  ë¡œê·¸ ì„¤ì •
logging.basicConfig(level=logging.INFO, format='[%(asctime)s] %(message)s')


def extract_tables_from_excel(file_path):
    """ì—‘ì…€ íŒŒì¼ì—ì„œ ì—¬ëŸ¬ í…Œì´ë¸”ì„ ìë™ ë¶„ë¦¬í•˜ì—¬ ì œëª©, ìš”ì•½, ë§ˆí¬ë‹¤ìš´ ë°˜í™˜"""
    xl = pd.ExcelFile(file_path)
    results = []

    for sheet_name in xl.sheet_names:
        logging.info(f" ì‹œíŠ¸ ì²˜ë¦¬ ì¤‘: {sheet_name}")
        df = xl.parse(sheet_name, header=None, dtype=str)

        current_table = []
        table_list = []

        #  ë¹ˆ í–‰ ê¸°ì¤€ìœ¼ë¡œ í…Œì´ë¸” ë¶„í• 
        for _, row in df.iterrows():
            if row.isnull().all():
                if current_table:
                    table_list.append(pd.DataFrame(current_table).reset_index(drop=True))
                    current_table = []
            else:
                current_table.append(row)
        if current_table:
            table_list.append(pd.DataFrame(current_table).reset_index(drop=True))

        logging.info(f"âœ… í…Œì´ë¸” {len(table_list)}ê°œ ë¶„í•  ì™„ë£Œ")

        # í…Œì´ë¸”ë³„ ë¶„ì„
        for i, table in enumerate(table_list):
            table_cleaned = table.dropna(axis=1, how='all')
            if table_cleaned.empty or table_cleaned.shape[0] < 2:
                continue

            # ì œëª© ì¶”ì¶œ
            first_row = table_cleaned.iloc[0].dropna().astype(str).tolist()
            title = " / ".join(first_row[:2]) if first_row else f"{sheet_name} Table {i+1}"

            #  í—¤ë” ì„¤ì •
            table_cleaned.columns = table_cleaned.iloc[0]
            data = table_cleaned[1:].copy()

            #  ìˆ˜ì¹˜ ìš”ì•½
            numeric_cols = data.apply(pd.to_numeric, errors='coerce').dropna(axis=1, how='all')
            summary = {}
            for col in numeric_cols.columns:
                col_data = numeric_cols[col].dropna()
                if not col_data.empty:
                    summary[col] = {
                        "max": col_data.max(),
                        "mean": col_data.mean()
                    }

            #  ë§ˆí¬ë‹¤ìš´ ì¶œë ¥
            markdown = data.to_markdown(index=False)

            results.append({
                "sheet": sheet_name,
                "title": title,
                "shape": data.shape,
                "summary": summary,
                "markdown": markdown
            })

    return results


def extract_images_from_excel(file_path):
    """ì—‘ì…€ íŒŒì¼ì—ì„œ ì‚½ì…ëœ ì´ë¯¸ì§€ ì¶”ì¶œ (base64ë¡œ ì¸ì½”ë”©ëœ PNG)"""
    wb = load_workbook(file_path)
    image_results = []

    for sheetname in wb.sheetnames:
        ws = wb[sheetname]
        if hasattr(ws, "_images") and ws._images:
            for idx, img in enumerate(ws._images):
                if hasattr(img, 'image'):
                    img_bytes = io.BytesIO()
                    img.image.save(img_bytes, format='PNG')
                    img_bytes.seek(0)
                    encoded = base64.b64encode(img_bytes.read()).decode()
                    image_results.append({
                        "sheet": sheetname,
                        "image_index": idx,
                        "base64": encoded
                    })

    logging.info(f" ì´ë¯¸ì§€ {len(image_results)}ê°œ ì¶”ì¶œ ì™„ë£Œ")
    return image_results


#  ì‹¤í–‰ ì˜ˆì‹œ
if __name__ == "__main__":
    file_path = r"C:\Ai_x\source\í”„ë¡œì íŠ¸2\sample_inputs\sample.xlsx"  # ë¶„ì„í•  ì—‘ì…€ íŒŒì¼ ê²½ë¡œ

    # 1. í‘œ/ë§ˆí¬ë‹¤ìš´/ìš”ì•½ ë¶„ì„
    tables = extract_tables_from_excel(file_path)
    for t in tables:
        print(f"\n###  {t['title']}")
        print(f"-  í¬ê¸°: {t['shape'][0]}í–‰ Ã— {t['shape'][1]}ì—´")
        if t["summary"]:
            print("-  ì£¼ìš” ìˆ˜ì¹˜ ìš”ì•½:")
            for col, stats in t["summary"].items():
                max_val = stats['max']
                mean_val = stats['mean']
                max_str = f"{max_val:.2f}" if isinstance(max_val, (int, float)) else str(max_val)
                mean_str = f"{mean_val:.2f}" if isinstance(mean_val, (int, float)) else str(mean_val)
                print(f"  - ì—´ '{col}': ìµœëŒ€={max_str}, í‰ê· ={mean_str}")
        print("\n" + t["markdown"])

    #  2. ì´ë¯¸ì§€ ì¶”ì¶œ
    images = extract_images_from_excel(file_path)
    for img in images:
        print(f"\n ì´ë¯¸ì§€ - ì‹œíŠ¸: {img['sheet']} / ì¸ë±ìŠ¤: {img['image_index']}")
        print(f"<img src='data:image/png;base64,{img['base64'][:80]}...'>")  # base64 ì¼ë¶€ ë¯¸ë¦¬ë³´ê¸°

```
